```python
import pandas as pd
df=messages = pd.read_csv('SMSSpamCollection', sep='\t', names=["label", "message"])
df.head()
```

```python
df.shape
```

```python
X=list(df['message'])
```

```python
y=list(df['label'])
```

```python
y
```

```python
y=list(pd.get_dummies(y,drop_first=True)['spam'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
!pip install transformers
```

```python
from transformers import DistilBertTokenizerFast
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
```

```python
train_encodings = tokenizer(X_train, truncation=True, padding=True)
test_encodings = tokenizer(X_test, truncation=True, padding=True)
```

```python
y_train
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test
))
```

```python
!pip install TFTrainer
```

```python
from tftrainer import TrainArgument, Trainer
from transformers import BertModel

training_args = TrainArgument(
    output_dir='./results',          # output directory
    num_train_epochs=2,              # total number of training epochs
    per_device_train_batch_size=8,  # batch size per device during training
    per_device_eval_batch_size=16,   # batch size for evaluation
    warmup_steps=500,                # number of warmup steps for learning rate scheduler
    weight_decay=0.01,               # strength of weight decay
    logging_dir='./logs',            # directory for storing logs
    logging_steps=10,
)
```

```python
with training_args.strategy.scope():
    model = BertModel.from_pretrained("bert-base-cased", use_safetensors=True)

trainer = TFTrainer(
    model=model,                    # the instantiated  Transformers model to be trained
    args=training_args,                  # training arguments, defined above
    train_dataset=train_dataset,         # training dataset
    eval_dataset=test_dataset             # evaluation dataset
)
     
trainer.train()
```

```python
trainer.evaluate(test_dataset)
```

```python
trainer.predict(test_dataset)
```

```python
output=trainer.predict(test_dataset)[1]
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,output)
cm
```
